```python
import pandas as pd
from difflib import SequenceMatcher

# Rutas a los archivos
archivo_1 = "00Tesis.xlsx"
archivo_2 = "CDEM.xlsx"

# Leer archivos Excel
df1 = pd.read_excel(archivo_1)
df2 = pd.read_excel(archivo_2)

# Verificar columnas necesarias
for df, name in [(df1, '00Tesis'), (df2, 'CDEM')]:
    if 'Titulo' not in df.columns or 'Fichero' not in df.columns:
        raise ValueError(f"El archivo '{name}' no tiene las columnas requeridas: 'Titulo' y 'Fichero'")

# Función para similitud
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

# Umbral de similitud para Título
umbral_similitud = 0.85

ficheros_diferentes = []

# Comparar títulos
for _, row1 in df1.iterrows():
    titulo1 = str(row1['Titulo']).strip()
    fichero1 = str(row1['Fichero']).strip() if pd.notna(row1['Fichero']) else ''
    
    mejor_similitud = 0
    mejor_titulo2 = ''
    fichero2 = ''
    
    for _, row2 in df2.iterrows():
        titulo2 = str(row2['Titulo']).strip()
        sim = similar(titulo1, titulo2)
        
        if sim > mejor_similitud:
            mejor_similitud = sim
            mejor_titulo2 = titulo2
            fichero2 = str(row2['Fichero']).strip() if pd.notna(row2['Fichero']) else ''
    
    if mejor_similitud >= umbral_similitud and fichero1 != fichero2:
        ficheros_diferentes.append((titulo1, mejor_titulo2, fichero1, fichero2))

# Mostrar resultados
print("\nFICHEROS DISTINTOS PARA TÍTULOS SIMILARES:")
for _, _, f1, f2 in ficheros_diferentes:
    print(f"- fichero1: '{f1}' vs fichero2: '{f2}'")

```
